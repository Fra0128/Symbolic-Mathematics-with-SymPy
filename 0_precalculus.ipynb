## Functions

*Definition*: A **function** is a correspondence between input numbers (x-values) and output numbers (y-values) that sends each input numbers (x-value) to exatly one output number (y-value).

Sometimes, a function is described with an equation 

Example: $$y = x^2 + 1$$ which can also be written as $$f(x) = x^2 + 1$$

- what is $f(2)$? $\rightarrow$ $f(2) = 2^2 + 1 = 5$
- what is $f(a + 3)$? $\rightarrow$ $f(a + 3) = (a + 3)^2 + 1 = a^2 + 9 + 6a + 1 = 10$

Sometimes, a function is described with a **graph**.

## Increasing and decreasing functions

*Definition*: **a function is increasing** when $x_2>x_1$, $\rightarrow$ $f(x_1)>f(x_2)$

**a function is decreasing** when $x_2>x_1$, $\rightarrow$ $f(x_1)<f(x_2)$



In general we can say that a function is increasing if x goes from left to right and y goes down.

*Definition*: A function $f(x)$ has an **absolute maximum** at x=c if the y-value at x=c is as big it ever get.

$f(c)>f(x)$ for all x-values in the domain of f.

the y-value $f(c)$ is called the **absolute maximum value** for f and the point $(c, f(c))$ is called the absolute maximum point.

*Definition*: A function $f(x)$ has an **absolute minimum** at x=c if the y-value at x=c is as smallest it ever get.

$f(c) \leq f(x)$ for all x-values in the domain of f.

the y-value $f(c)$ is called the **absolute minimum value** for f and the point $(c, f(c))$ is called the absolute minimum point.


*Definition*: A function $f(x)$ has a **local maximum value** if $f(c) \ge f(x)$ for all x-values in a open interval around c.

the y-value $f(c)$ is called the **local maximum value** for f and the point $(c, f(c))$ is called the local maximum point.

*Definition*: A function $f(x)$ has a **local minimum value** if $f(c) \le f(x)$ for all x-values in a open interval around c.

the y-value $f(c)$ is called the **local minimum value** for f and the point $(c, f(c))$ is called the local minimum point.



*Definition*: local maximum and minimum  can also be called relative maximum and minimum.

## Even and odd functions

a function is **even** if its graph is symmetric with respect to the y-axis and f(-x) = f(x) for all x in the domain of f

a function is **odd** if its graph is symmetric with respect to the origin and f(-x) = -f(x) for all x in the domain of f

There is no definition for functions that are symmetric with x-axis because it is not a function

## Transformers of functions

### Rules for Trasformations

**Numbers** on the **outside** of the function affect the **y-values** and result in *vertical motions*.
This motions are in the direction yuo expect.

**Numbers** on the **inside** of the function affect the **x-values** and result in *horizontal motions*.
This motions go in the opposite direction from what you expect.

- **Adding** results in a **shift** (translation)
- **Multiplying** results in a **stretch or shrink**
- A **negative sign** resuts in a **reflection**

#### Sympy

```python
# define the symbols and functions
x = symbols('x')
y1 = sqrt(x)
y2 = sqrt(x) - 2
y3 = sqrt(x - 2)
y4 = 3 * sqrt(x)
y5 = -sqrt(x)
y6 = sqrt(x/4)
# plot the functions on the same set of axes
p = plotting.plot(y1, y2, y3, y4, y5, y6, show = False, title = 'Example of Transformers of functions')

# add the legend
p[0].label = '$y = \sqrt{x}$'
p[1].label = '$y = \sqrt{x}-2$'
p[2].label = '$y = \sqrt{x-2}$'
p[3].label = '$y = 3 \\dot \sqrt{x}$'
p[4].label = '$y = - \sqrt{x}$'
p[5].label = '$y = \sqrt{\\frac{x}{4}}$'
p.legend = True

# show the plot
p.show()
```

## Inverse Functions

*Definition*: An inverse function undoes what a function does

*Example*:

$f(x) = x + 2$

$f^{-1}(x) = x - 2$

**key fact (1):** Inverse functions *reverse* the roles of y and x!!!

**key fact (2):** the grapfh of $y=f^{-1}(x)$ is obtained from the graph of $y=f(x)$ by reflecting over the line $y=x$.



**key fact (3):** the original function $f(x)$ and $f^{-1}(x)$ are both rational, BUT they are not the reciprocal of each other:

$$f^{-1}(x) \neq \frac{1}{f(x)}$$

Do all functions have inverse functions? -> NO!!!



**KEY FACT (4)** a function $f$ has an inverse function in and only if the graph of $f$ satisfies the horizontal line test



**key fact (5)** for any invertible function $f$, the domai of $f^{-1}$ is the range of $f(x)$ and the range of $f^{-1}$ is the domain of $f(x)$

This rules will be important when we study exponential functions and their inverses logarithmic functions

#### SymPy

```python
f = x**2
fInv = Symbol("f^-1")           # define a symbol for the inverse
inv = solve(f * fInv - 1, fInv) # solve for f * fInv == 1
inv
```

## Angles and Their Measures

There are different ways of describing the size of an angle. Angles are commonly measured either degrees or radiants.

For degrees: a full circle measures 360° for convention

For radiants: a full circle measures $2\pi$

*Note*. To convert between degrees and radiants, it is handy to use the fact that 180° equals $\pi$ radians.

*Example:* Convert -135° to radiants

$$\frac{-135°}{180°} = \frac{x}{\pi \cdot radians}$$

$$x = -0.75 \cdot \pi \cdot radians$$

$$x = -2.3562 \cdot radians$$

Sometimes, angles are given in terms of degrees, minutes and seconds as in 32°17'25''.

- A minute is $\frac{1}{60}$ degree -> $60 \; min = 1 \; degree$
- A second is $\frac{1}{60}$ minute -> $\frac{1}{60}$ $\frac{1}{60}$ degree = $\frac{1}{3600}$ degree -> $3600 \; sec  = 1 \; degree$


## Arclength and Areas of Sectors

The circumgerence of a circle is given by the formula: $$C = 2 \pi r$$

*Example*: A circular pool has a radius of 8m. Find the arclength spanned by a central angle of 2.5 radians.

$C = 2 \pi 8m$

$arclength = fraction \; of \;  circle \cdot circumference$

$\frac{2.5 \; radians}{2\pi \; radians} \cdot 2 \pi 8m = 20m$ 

The arclength is related to the angle it spans according to the formul: 

$$arclength = \frac{\theta}{2 \pi} \cdot 2\pi r = \theta \cdot r$$

The area of a circle is given by the formula: $$A= \pi r^2$$

*Ecample*: find the area of a sector of a circle of radians of 10m, that spans an angle of $\frac{\pi}{6}$ rad

$A = \pi 10^2 = 100 \pi \;m^2$

$area of sector = fraction of circle \cdot Area of circle$

$$\frac{\frac{\pi}{6} \; radians}{2 \pi \; radians} \cdot 100 \pi \; m^2 = 26.18 \; m^2$$


The area od a sector is related to the angle is spans according to the form la: $$Area \; of \; sector = \frac{\theta}{2 \pi} \cdot \pi r^2 = \frac{\theta}{2} \cdot r^2$$
