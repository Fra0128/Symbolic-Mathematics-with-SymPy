## Functions

*Definition*: A **function** is a correspondence between input numbers (x-values) and output numbers (y-values) that sends each input numbers (x-value) to exatly one output number (y-value).

Sometimes, a function is described with an equation 

Example: $$y = x^2 + 1$$ which can also be written as $$f(x) = x^2 + 1$$

- what is $f(2)$? $\rightarrow$ $f(2) = 2^2 + 1 = 5$
- what is $f(a + 3)$? $\rightarrow$ $f(a + 3) = (a + 3)^2 + 1 = a^2 + 9 + 6a + 1 = 10$

Sometimes, a function is described with a **graph**.

```python
y = x**2 + 1
plot(y, size=(3, 3), title='$y = x^2$', ylim = (0, 50))
```

### Domain and range of a function



In its simplest form the **domain** is all the values that go into a function, and the **range** is all the values that come out.

```python
display('The doamin of our function y is ', calculus.util.continuous_domain(y, x, S.Reals)) 
display('The range of our function y is ', calculus.util.function_range(y, x, S.Reals))
```

What are the domain domain of these functions?

A. $$f(x) = \frac{x}{x^2 - 4x + 3}$$ 

B. $$f(x) = \sqrt{3 - 2x}$$

C. $$f(x) = \frac{\sqrt{3 - 2x}}{x^2 - 4x + 3}$$

```python
A = x / ( x**2 - 4*x + 3)
B = sqrt(3 - 2*x)
C = sqrt(3 - 2*x) / (x**2 - 4*x + 3)

display('A', calculus.util.continuous_domain(A, x, S.Reals))
display('B', calculus.util.continuous_domain(B, x, S.Reals))
display('C', calculus.util.continuous_domain(C, x, S.Reals))
```

to obtain these results:
- A. $\rightarrow$ the denominator has to be $\neq 0$
- B. $\rightarrow$ the number inside the square root has to be $\geq 0$
- C. $\rightarrow$ here the 2 previous rules have to be respected at the same time

Note: the "[ ]" are used to show that after them the domain doesn't exist
